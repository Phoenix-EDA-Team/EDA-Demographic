```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. 데이터 불러오기 및 헤더 처리
# --------------------------------------------------
# 헤더가 3줄(0,1,2)로 구성된 경우
df = pd.read_csv('./data/인구순이동률2012-2022-연령별.csv', header=[0,1,2], index_col=0)

# 열 이름 확인 (디버깅용)
print("=== Columns Structure ===")
print(df.columns.values)
```

```python
# 2. 열 구조 재정의
# --------------------------------------------------
# 기존 열 구조를 평탄화
df.columns = ['_'.join(col).strip() for col in df.columns.values]

# 새 열 이름 확인
print("\n=== Flattened Columns ===")
print(df.columns.tolist()[:10])  # 첫 10개 열만 출력
```

```python
# 3. 청년층(20-35세) 열 선택
# --------------------------------------------------
# 수정된 연령대 패턴 (공백 포함)
target_ages = [
    '20 - 24세', 
    '25 - 29세', 
    '30 - 34세', 
]

# 모든 성별(남자/여자)과 지역 포함
target_columns = [
    col for col in df.columns 
    if any(age in col for age in target_ages)
]

print("\n=== 청년층 열 목록 (첫 10개) ===")
print(target_columns[:10])
```

```python
# 4. 지역 분류
# --------------------------------------------------
수도권_지역 = ['서울특별시', '인천광역시', '경기도']
지방_지역 = ['부산광역시', '대구광역시', '광주광역시', '대전광역시', '울산광역시',
          '강원특별자치도', '충청북도', '충청남도', '전북특별자치도', 
          '전라남도', '경상북도', '경상남도', '제주특별자치도']

# 수도권 데이터 추출
수도권_열 = [col for col in target_columns if col.split('_')[-1] in 수도권_지역]
수도권_데이터 = df[수도권_열].mean(axis=1)

# 지방 데이터 추출
지방_열 = [col for col in target_columns if col.split('_')[-1] in 지방_지역]
지방_데이터 = df[지방_열].mean(axis=1)

# 결과 데이터프레임
result_df = pd.DataFrame({
    '수도권': 수도권_데이터,
    '지방': 지방_데이터
}, index=df.index)
```

```python
# 데이터프레임을 기존 데이터로 가정하고 시작합니다. 예: df_raw = pd.read_csv('your_file.csv')
# 예제에서는 df_raw가 이미 DataFrame 형태라고 가정합니다.

# 연령대 정의
age_groups = {
    '어린이': [f'0 - 4세.{i}' for i in range(9)] + [f'5 - 9세.{i}' for i in range(9)] + [f'10 - 14세.{i}' for i in range(5)],
    '청소년': [f'10 - 14세.{i}' for i in range(5,9)] + [f'15 - 19세.{i}' for i in range(5)],
    '청년': [f'15 - 19세.{i}' for i in range(5,9)] + [f'20 - 24세.{i}' for i in range(9)] +
            [f'25 - 29세.{i}' for i in range(9)] + [f'30 - 34세.{i}' for i in range(9)],
    '중년': [f'35 - 39세.{i}' for i in range(9)] + [f'40 - 44세.{i}' for i in range(9)] +
            [f'45 - 49세.{i}' for i in range(9)] + [f'50 - 54세.{i}' for i in range(9)] +
            [f'55 - 59세.{i}' for i in range(9)] + [f'60 - 64세.{i}' for i in range(9)],
    '노인': [f'65 - 69세.{i}' for i in range(9)] + [f'70 - 74세.{i}' for i in range(9)] +
            [f'75 - 79세.{i}' for i in range(9)] + [f'80세이상.{i}' for i in range(9)]
}

# 원본에서 필요한 데이터만 추출 (2012~2022, index 3~13), '전국'부터 '제주특별자치도'까지
years = list(range(2012, 2023))
regions = ['전국', '서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시',
           '대전광역시', '울산광역시', '세종특별자치시', '경기도', '강원특별자치도',
           '충청북도', '충청남도', '전북특별자치도', '전라남도', '경상북도', '경상남도', '제주특별자치도']

# 필요한 컬럼 인덱스를 파악
region_row = df_raw.iloc[1]
columns_needed = [i for i, val in enumerate(region_row) if val in regions]

# 결과 저장 리스트
records = []

for idx, year in enumerate(years):
    row_index = idx + 3  # 연도 데이터는 4번째 행부터 시작
    for col_idx in columns_needed:
        region = df_raw.iloc[1, col_idx]  # 지역명 가져오기

        for group, colnames in age_groups.items():
            group_cols = [col for col in colnames if col in df_raw.columns]

            # 💡 값 가져와서 숫자로 변환
            migration_rates = []
            for col in group_cols:
                value = df_raw.loc[row_index, col]
                try:
                    value = float(value)
                    migration_rates.append(value)
                except (ValueError, TypeError):
                    continue  # 변환 안 되면 건너뜀

            # 평균 계산
            if migration_rates:
                mean_migration = sum(migration_rates) / len(migration_rates)
                records.append({
                    'Year': year,
                    'Region': region,
                    'AgeGroup': group,
                    'MigrationRate': mean_migration
                })

# 최종 DataFrame 생성
df_result = pd.DataFrame(records)

# 예시 출력
print(df_result.head())
```

```python
import  seaborn as sns
import matplotlib.pyplot as plt

# 한글폰트 설정 2
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname='./resources/fonts/NotoSansKR-Regular.otf').get_name()
rc('font',family=font_name)
# 한글폰트 설정 1
plt.rcParams['font.family'] = 'Noto Sans KR'
```

```python
# 연도별, 수도권/비수도권별 순이동률 평균 계산
pivot = capital_yuth.groupby(['Year', 'Region_group'])['MigrationRate'].mean().reset_index()

# 피벗테이블로 변환 (그래프 그리기 편하게)
pivot_table = pivot.pivot(index='Year', columns='Region_group', values='MigrationRate')

# 그래프 그리기
plt.figure(figsize=(10,6))
pivot_table.plot(marker='o')
plt.title('청년층 수도권 vs 비수도권 연도별 순이동률 변화')
plt.ylabel('평균 순이동률')
plt.xlabel('Year')
plt.grid(True)
plt.legend(title='지역구분')
plt.tight_layout()
plt.show()
```
