```python
# ë°ì´í„°í”„ë ˆì„ì„ ê¸°ì¡´ ë°ì´í„°ë¡œ ê°€ì •í•˜ê³  ì‹œì‘í•©ë‹ˆë‹¤. ì˜ˆ: df_raw = pd.read_csv('your_file.csv')
# ì˜ˆì œì—ì„œëŠ” df_rawê°€ ì´ë¯¸ DataFrame í˜•íƒœë¼ê³  ê°€ì •í•©ë‹ˆë‹¤.

# ì—°ë ¹ëŒ€ ì •ì˜
age_groups = {
    'ì–´ë¦°ì´': [f'0 - 4ì„¸.{i}' for i in range(9)] + [f'5 - 9ì„¸.{i}' for i in range(9)] + [f'10 - 14ì„¸.{i}' for i in range(5)],
    'ì²­ì†Œë…„': [f'10 - 14ì„¸.{i}' for i in range(5,9)] + [f'15 - 19ì„¸.{i}' for i in range(5)],
    'ì²­ë…„': [f'15 - 19ì„¸.{i}' for i in range(5,9)] + [f'20 - 24ì„¸.{i}' for i in range(9)] +
            [f'25 - 29ì„¸.{i}' for i in range(9)] + [f'30 - 34ì„¸.{i}' for i in range(9)],
    'ì¤‘ë…„': [f'35 - 39ì„¸.{i}' for i in range(9)] + [f'40 - 44ì„¸.{i}' for i in range(9)] +
            [f'45 - 49ì„¸.{i}' for i in range(9)] + [f'50 - 54ì„¸.{i}' for i in range(9)] +
            [f'55 - 59ì„¸.{i}' for i in range(9)] + [f'60 - 64ì„¸.{i}' for i in range(9)],
    'ë…¸ì¸': [f'65 - 69ì„¸.{i}' for i in range(9)] + [f'70 - 74ì„¸.{i}' for i in range(9)] +
            [f'75 - 79ì„¸.{i}' for i in range(9)] + [f'80ì„¸ì´ìƒ.{i}' for i in range(9)]
}

# ì›ë³¸ì—ì„œ í•„ìš”í•œ ë°ì´í„°ë§Œ ì¶”ì¶œ (2012~2022, index 3~13), 'ì „êµ­'ë¶€í„° 'ì œì£¼íŠ¹ë³„ìì¹˜ë„'ê¹Œì§€
years = list(range(2012, 2023))
regions = ['ì „êµ­', 'ì„œìš¸íŠ¹ë³„ì‹œ', 'ë¶€ì‚°ê´‘ì—­ì‹œ', 'ëŒ€êµ¬ê´‘ì—­ì‹œ', 'ì¸ì²œê´‘ì—­ì‹œ', 'ê´‘ì£¼ê´‘ì—­ì‹œ',
           'ëŒ€ì „ê´‘ì—­ì‹œ', 'ìš¸ì‚°ê´‘ì—­ì‹œ', 'ì„¸ì¢…íŠ¹ë³„ìì¹˜ì‹œ', 'ê²½ê¸°ë„', 'ê°•ì›íŠ¹ë³„ìì¹˜ë„',
           'ì¶©ì²­ë¶ë„', 'ì¶©ì²­ë‚¨ë„', 'ì „ë¶íŠ¹ë³„ìì¹˜ë„', 'ì „ë¼ë‚¨ë„', 'ê²½ìƒë¶ë„', 'ê²½ìƒë‚¨ë„', 'ì œì£¼íŠ¹ë³„ìì¹˜ë„']

# í•„ìš”í•œ ì»¬ëŸ¼ ì¸ë±ìŠ¤ë¥¼ íŒŒì•…
region_row = df_raw.iloc[1]
columns_needed = [i for i, val in enumerate(region_row) if val in regions]

# ê²°ê³¼ ì €ì¥ ë¦¬ìŠ¤íŠ¸
records = []

for idx, year in enumerate(years):
    row_index = idx + 3  # ì—°ë„ ë°ì´í„°ëŠ” 4ë²ˆì§¸ í–‰ë¶€í„° ì‹œì‘
    for col_idx in columns_needed:
        region = df_raw.iloc[1, col_idx]  # ì§€ì—­ëª… ê°€ì ¸ì˜¤ê¸°

        for group, colnames in age_groups.items():
            group_cols = [col for col in colnames if col in df_raw.columns]

            # ğŸ’¡ ê°’ ê°€ì ¸ì™€ì„œ ìˆ«ìë¡œ ë³€í™˜
            migration_rates = []
            for col in group_cols:
                value = df_raw.loc[row_index, col]
                try:
                    value = float(value)
                    migration_rates.append(value)
                except (ValueError, TypeError):
                    continue  # ë³€í™˜ ì•ˆ ë˜ë©´ ê±´ë„ˆëœ€

            # í‰ê·  ê³„ì‚°
            if migration_rates:
                mean_migration = sum(migration_rates) / len(migration_rates)
                records.append({
                    'Year': year,
                    'Region': region,
                    'AgeGroup': group,
                    'MigrationRate': mean_migration
                })

# ìµœì¢… DataFrame ìƒì„±
df_result = pd.DataFrame(records)

# ì˜ˆì‹œ ì¶œë ¥
print(df_result.head())
```
