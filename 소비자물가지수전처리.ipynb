```python
import pandas as pd
```

```python
df = pd.read_csv('./data/소비자물가지수2002-2022.csv',encoding='utf-8')

df.info()
```

```python
# '주요품목별(1)' 열을 인덱스로 설정
df.set_index("주요품목별(1)", inplace=True)
df.head(3)
```

```python
df = df.apply(pd.to_numeric, errors="coerce")

# 숫자형 열만 선택
numeric_columns = df.select_dtypes(include=["float", "int"])

# 연도별 평균 계산
yearly_avg = numeric_columns.mean(axis=0).reset_index()

# 새로운 DataFrame 생성
yearly_avg.columns = ["Year", "Average_Price"]

# yearly_avg.drop(yearly_avg.iloc[0])

# 결과 출력
display(yearly_avg)

# 
```

```python
#데이터 저장
yearly_avg.to_csv('./data/수도권물가평균2002-2022.csv',index=False)
```
