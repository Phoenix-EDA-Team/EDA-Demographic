```python
import pandas as pd
```

```python
df = pd.read_csv('./data/지역별1인당소득2002-2022.csv',encoding='utf-8')
df.drop(columns=['항목'],inplace=True) 
```

```python
df['년도'] = df['시점'].rename('년도',inplace=True)
```

```python
df["년도"] = df["년도"].replace("2022 p)", "2022")
df.drop(columns=['시점'],inplace=True)
df
```

```python
# 데이터프레임을 변환하여 지역 컬럼 추가
df_melted = pd.melt(df,
                    id_vars=["년도"],
                    var_name="지역",
                    value_name="1인당 개인소득")
df_melted.head(50)
```

```python
df_melted = df_melted[df_melted['지역'] != '항목']
df_melted
```

```python
# 데이터프레임을 변환하여 지역 컬럼 추가
df_melted = pd.melt(df,
                    id_vars=["년도"],
                    var_name="지역",
                    value_name="1인당 개인소득")

# 지역 이름을 영어로 변환
region_map = {
    "전국" : "Whole country",
    "서울특별시": "Seoul",
    "부산광역시": "Busan",
    "대구광역시": "Daegu",
    "인천광역시": "Incheon",
    "광주광역시": "Gwangju",
    "대전광역시": "Daejeon",
    "울산광역시": "Ulsan",
    "세종특별자치시": "Sejong",
    "경기도": "Gyeonggi-do",
    "강원도": "Gangwon-do",
    "충청북도": "Chungcheongbuk-do",
    "충청남도": "Chungcheongnam-do",
    "전라북도": "Jeollabuk-do",
    "전라남도": "Jeollanam-do",
    "경상북도": "Gyeongsangbuk-do",
    "경상남도": "Gyeongsangnam-do",
    "제주특별자치도": "Jeju"
}

df_melted["지역"] = df_melted["지역"].map(region_map)
# df_melted = df_melted.rename(columns={"년도": "Year"})
# df_melted = df_melted.rename(columns={"지역": "Region"})


# 결과 출력
display(df_melted.head(50))

# 저장
df_melted.to_csv('./data/지역별1인당경제소득.csv',encoding='utf-8',index=False)
```
