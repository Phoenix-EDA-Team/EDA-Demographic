```python
import pandas as pd
```

```python
df = pd.read_csv('./data/인구전입률2012-2022.csv')
```

```python
df
```

```python
# 기존 열 이름 리스트 (시점 제외)
columns = ['전국', '서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시', '울산광역시', '세종특별자치시', '경기도', '강원특별자치도', '충청북도', '충청남도', '전북특별자치도', '전라남도', '경상북도', '경상남도', '제주특별자치도']

# '순이동률(%)'에서 % 제거하고 '순이동률'만 붙이기
new_columns = [f"{col}" for col in columns]

# 데이터프레임 열 이름 변경 (예: df가 원본 데이터프레임)
df.columns = ['Year'] + new_columns
```

```python
df = df.drop(index=0).reset_index(drop=True)  # 첫 번째(0번) 행 제거 후 인덱스 재정렬
df
```

```python
df_long = df.melt(id_vars='Year', var_name='지역', value_name='전입률')

# 컬럼명 변경
df_long = df_long.rename(columns={'Year': '년도'})

# 연도, 지역, 순이동률 순서로 정렬
df_long = df_long[['년도', '지역', '전입률']]

# 연도, 지역 기준 정렬
df_long = df_long.sort_values(['년도', '지역']).reset_index(drop=True)


df_long = df_long.rename(columns={'년도': 'Year'})
df_long = df_long.rename(columns={'지역': 'Region'})


display(df_long)
```

```python
# 지역 이름을 영어로 변환
region_map = {
    "전국" : "Whole country",
    "서울특별시": "Seoul",
    "부산광역시": "Busan",
    "대구광역시": "Daegu",
    "인천광역시": "Incheon",
    "광주광역시": "Gwangju",
    "대전광역시": "Daejeon",
    "울산광역시": "Ulsan",
    "세종특별자치시": "Sejong",
    "경기도": "Gyeonggi-do",
    "강원특별자치도": "Gangwon-do",
    "충청북도": "Chungcheongbuk-do",
    "충청남도": "Chungcheongnam-do",
    "전라북도": "Jeollabuk-do",
    "전라남도": "Jeollanam-do",
    "경상북도": "Gyeongsangbuk-do",
    "경상남도": "Gyeongsangnam-do",
    "제주특별자치도": "Jeju"
}

df_long["Region"] = df_long["Region"].map(region_map)



# 결과 출력
display(df_long.head(50))

# 저장
df_long.to_csv('./data/지역별전입률2012-2022.csv',encoding='utf-8',index=False)
```
